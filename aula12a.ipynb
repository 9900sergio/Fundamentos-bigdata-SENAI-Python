 ## Exercício 24 – 
Uma empresa tem um conjunto de dados sobre a dimensão de um
determinado produto feito com matéria-prima de três fornecedores. Você foi designado
a desenvolver uma análise para conseguir analisar quais desses fornecedores tem a
matéria prima que não garante o dimensional final deste produto, a media máxima é de
120,15 mm e a dimensão mínima é de 119,9 mm. Crie um histograma e um boxplot para
analisar as medidas de cada fornecedor e um boxplot para comparar os três
fornecedores. Utilize o arquivo fornecedores.xlsx

```python
## importar bibliotecas

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

## carregar o arquivo e inserir em um dataframe ctrl + F

path = 'C:\\Users\\FIC\\Documents\\Pedrosp\\003 - Arquivos\\' 
file ='fornecedores.xlsx'

#carregar o arquivo e inserir  em um dataframe

df = pd.read_excel(path + file)

# visualizar o cabeçalho do dataframe
df.head()

## tratamento dos dados. utilizar metodo iloc para excluir as tres primeiras linhas e a coluna unnamed:0

df = df.iloc[3:,1:]

df.columns = ['f_1','f_2','f_3']
df.head()

# construção histograma 1

plt.hist(df['f_1'])  # com matplotlib

sns.displot(df['f_1'], bins = 10) #com seaborn, bins determina o numero de classes (colunas)


# descobrir a forma da dataframe - retorna o numero de linhas e colunas do dataframe
#df.shape

#df.describe()

#verificar se ha dados nulos
#df.isnull()

#df.info()

```

```python
# construção histograma 2

plt.hist(df['f_2'])  # com matplotlib

sns.displot(df['f_2'], bins = 10) #com seaborn, bins determina o numero de classes (colunas)

```

```python
# construção histograma 3

plt.hist(df['f_3'])  # com matplotlib

sns.displot(df['f_3'], bins = 10) #com seaborn, bins determina o numero de classes (colunas)
```

```python
#Box plot - 

#converter objeto em numérico

df['f_1'] = pd.to_numeric(df['f_1'])
df['f_2'] = pd.to_numeric(df['f_2'])
df['f_3'] = pd.to_numeric(df['f_3'])

df.info()

sns.boxplot(y=df['f_1'])



```
