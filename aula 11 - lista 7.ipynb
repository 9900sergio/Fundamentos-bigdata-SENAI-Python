## Escreva um programa Python para desenhar uma linha com rótulo adequado no eixo x, eixo y e um título.Amostra de dados: x = range(1,50) y = [value * 3 for value in x]

```python
import matplotlib.pyplot as plt

x = range (1,50)
y = [value * 3 for value in x]

print('Valores de X')
print(range(1,50))
print('Valores de Y (Três vezes X)')
print (y)

plt.plot(x,y, marker='*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title ('Draw a line.')

```

## Escreva um programa Python para desenhar uma linha usando
# determinados valores de eixo com rótulo adequado no eixo x, eixo y e um título.
# x axis values
# x = [1,2,3]
# y axis values
# y = [2,4,1]


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3])
y = np.array([2,4,1])

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title ('Sample graph!')
```
