```python
# criando grafico de colunas

graficos = sns.barplot(x='Categoria', y= 'ValorVenda', data=df, orient ='v', errorbar=None) #orient = orientação

#metodo sns.barplot vai retornar uma lista de sub 
#metodos utilizando os caontainers para acessar o texto
#dos rotulos para cada coluna, desta forma o python entende
#que ele deve adicionar um rotulo em cada coluna, de acordo com o
#comando graficos.containers[0]

graficos.bar_label(graficos.containers[0])
```
