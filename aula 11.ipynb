```python
##importando biblioteca

import matplotlib.pyplot as plt ## permite gerar graficos
import numpy as np ## permite trabalhar com matrizes e vetores
```

```python
## primeiro grafico com matplotlib
## criando vetores - conjunto de dados
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

## plotar o grafico

plt.plot(x,y)


```

```python
## modificando o tipo de marcadores

plt.plot(x,y,marker='*')
```

```python
## alterar cor da linha

plt.plot(x,y,marker='*',c='b')
```

```python
## mudando como os pontos são ligados e o tipo de linha

plt.plot(
    x,y, ## os dados do grafico
    marker = '*', # tipo marcador
    c='green', # cor da linha
    ds = 'steps', #forma com que os pontos são conectados
    ls = '--' # estilo de linha
)
```

```python
## mudando cor dos pontos

plt.plot(
    x,y, ## os dados do grafico
    marker = '*', # tipo marcador
    c='green', # cor da linha
    ds = 'steps', #forma com que os pontos são conectados
    ls = '--', # estilo de linha
    mec = 'blue' #cor dos marcadores
)
```

```python
## matplotlib adicionando outros elementos

plt.plot(x,y) ## adiciona dados ao grafico

##adicionar titulo aos eixos x e y

plt.xlabel('valores de X', fontsize = '12')


plt.ylabel('valores de Y', fontsize = '12')

## adicionar titulo ao grafico

plt.title('Titulo do grafico', fontsize = '18', c = 'purple')

## mudando cor dos valores dos eixos
plt.xticks(fontsize = 8, c='blue')

plt.yticks(fontsize = 8, c='orange')

```

```python
import matplotlib.pyplot as plt
import numpy as np

# retorna Linspace valores igualmente espaçados em um intervalo
x = np.linspace (0,2*np.pi,100)
y = np.sin(x)
z = np.cos(x)
plt.plot (x,y)
plt.plot (x,z)
```

```python
import matplotlib.pyplot as plt
import numpy as np

## O comendoplt.subplot permite crir grupo de graficos
## plt.subplt ( linhas, colunas, posição elemento)

plt.subplot(2,1,1) # primeiro grafico
plt.plot(x,y)  # conjunto de dados do primeiro grafico
plt.title('subplots')
plt.subplot(2,1,2) #segundo grafico
plt.plot(x,z) # conjunto de dados do segundo grafico
```

## Seaborn Pairplot (bivariada)

```python
sns.pairplot(
    iris, #dataset ou conjunto de dados
    hue ='species', # legenda
    diag_kind = 'kde', #tipo de grafico de distrib
    kind = 'scatter', # tipo de grafico de correlação
    palette = 'husl'#conjunto de cores
)
plt.savefig('pairplot_iris.png')
```
